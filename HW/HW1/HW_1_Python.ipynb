‚úçüèº **Problem 3**: Write a paragraph about John W. Tukey. You can use the class sides, but also Wikipedia or any other source. Why is he important to data science? Open a markdown cell below this one to put your discussion.

John W. Tukey, who is considered the originator of Exploratory Data Analysis, published "Exploratory Data Analysis" at a time when computer-aided visualization was still a very new idea. He pioneered many of the key foundations of what came later to be known as Data Science. During the 1960s, Tukey challenged the dominance at the time of what he called "confirmatory data analysis", statistical analyses driven by rigid mathematical configurations. Tukey emphasized the importance of having a more flexible attitude towards data analysis and of exploring data carefully to see what structures and information might be contained therein. He called this ‚Äúexploratory data analysis‚Äù. In many ways, EDA was a precursor to Data Science.

‚úçüèº **Problem 6**: Describe in your own words, in a new markdown cell, what a tuple, a list and an array are in Python. Why do we need all three of these? How are they similar? How are they different? What are the use cases for each?

A list is a mutable ordered collection in python which is widely used to store data. 
A tuple is an immutable ordered collection.
A numpy array is a mutable ordered colection which can be imported to be used in python.

All of these have different features having different usecases and can even be used with each other.

Similarity: They all are ordered collections which support indexing.

Diferences: List and tuple are built in data structures while array has to be imported.
            List and array are mutable while tuple is not
            Array has to be declared which is not the case with list and tuple

Use cases: -
List: When we want to create a collection of elements which will store elements in insertion order and allow indexing to access or change data.
Tuple: When we want to create a collection of elements which will store elements in insertion order and allow indexing to access the data but the data should not be mutated.
Array: When we want to create an ordered collection of elements to store them compactly and perform numerical operations on them.
